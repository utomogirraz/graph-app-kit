```python
%matplotlib inline
from sklearn.manifold import TSNE
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from node2vec import Node2Vec
sns.set_style('whitegrid')
import pandas as pd
import warnings
```

```python
g_data = nx.from_pandas_edgelist(df, source='src',target='dest')
```

```python
#Ingredient
ing = df[['src','dest']].dropna(axis = 0,how = 'any')
ing.head()
```

```python
mylist = ing.values.tolist()
mylist
```

```python
g_data.add_edges_from(mylist,weight=1,label="Ingredient")
```

```python
labels = [i for i in dict(g_data.nodes).keys()]
labels = {i:i for i in dict(g_data.nodes).keys()}
labels
```

```python
node2vec = Node2Vec(g_data, dimensions=100, walk_length=16, num_walks=5)
```

```python
model = node2vec.fit(window=10,min_count=1)
```

```python
nodeFood =[x for x in df.src]
nodeFood
```

```python
embeddings = np.array([model.wv[x] for x in nodeFood])
embeddings
```

```python
model.wv.save_word2vec_format('foodingredient.emb')
```

```python
words = list(model.wv.vocab)
print(words)
```

```python
model.wv.save_word2vec_format('foodingredient.emb','vocab.txt')
```

```python
tsne = TSNE(n_components=2, random_state=7, perplexity=15)
embeddings_2d = tsne.fit_transform(embeddings)
```

```python
import streamlit as st
figure = plt.figure()

ax = figure.add_subplot(111)

ax.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1])
st.write(figure)
```
